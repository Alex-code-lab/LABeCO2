```python
import pandas as pd

# Chemin du fichier Excel
excel_path = "/Users/souchaud/Documents/Travail/Transition_ecologique/LABeCO2/data_masse_eCO2/code_NACRES/nacres-janvier-2022.xls"

# Lecture des colonnes "Codes Nacres" et "Libellé étendu nouvelle nomenclature achat"
df = pd.read_excel(excel_path, usecols=["Codes Nacres", "Libellé étendu nouvelle nomenclature achat"])

# Nettoyage des données : suppression des lignes où une des colonnes serait manquante
df = df.dropna(subset=["Codes Nacres", "Libellé étendu nouvelle nomenclature achat"])

# Chemin de sortie pour le fichier HDF5
hdf5_path = "/Users/souchaud/Documents/Travail/Transition_ecologique/LABeCO2/data_masse_eCO2/nacres_2022.h5"

# Sauvegarde du DataFrame dans le fichier HDF5 sous la clé "nacres_2022"
df.to_hdf(hdf5_path, key="nacres_2022", mode="w")

print("Fichier HDF5 'nacres_2022.h5' créé avec succès.")
```
