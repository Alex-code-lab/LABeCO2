```python
CNRS = CNRS.rename(columns={'Total consommé par code NACRES et fournisseur': 'Montant'})
CNRS
```

```python
ENS

```

```python
tableau_final = pd.concat([CNRS, INSERM, ENS], ignore_index=True)
tableau_final
```

```python
tableau_final['Montant'] = tableau_final['Montant'].str.replace(',', '.')
tableau_final['Montant'] = tableau_final['Montant'].str.replace(' ', '').astype(float)
tableau_final
```

```python
tableau_final_agrege = tableau_final.groupby('Code NACRES')['Montant'].sum().reset_index()
```

```python
tableau_final_agrege = tableau_final_agrege.sort_values('Montant', ascending=False).reset_index()
tableau_final_agrege.head(10)
```

```python
import matplotlib.pyplot as plt
premiers_50 = tableau_final_agrege[0:100]
plt.figure(figsize=(15,6))
plt.bar(premiers_50['Code NACRES'], premiers_50['Montant'])
plt.xticks(rotation=45, ha='right')
plt.title('Histogramme des 50 premiers Codes NACRES')
plt.xlabel('Code NACRES')
plt.ylabel('Montant')
plt.tight_layout()
plt.show()
```
